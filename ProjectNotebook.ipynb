I decided to implement the game 2048 (http://2048game.com/). The object of the game is to reach the 2048 tile by sliding the tiles left, right, up, or down and combining tiles of the same number values (powers of 2's: 2,4,8,..).

Description from Wikepedia(https://en.wikipedia.org/wiki/2048_(video_game)):
> 2048 is played on a gray 4Ã—4 grid, with numbered tiles that slide smoothly when a player moves them using the four arrow keys. Every turn, a new tile will randomly appear in an empty spot on the board with a value of either 2 or 4. Tiles slide as far as possible in the chosen direction until they are stopped by either another tile or the edge of the grid. If two tiles of the same number collide while moving, they will merge into a tile with the total value of the two tiles that collided. The resulting tile cannot merge with another tile again in the same move. Higher-scoring tiles emit a soft glow.

> A scoreboard on the upper-right keeps track of the user's score. The user's score starts at zero, and is increased whenever two tiles combine, by the value of the new tile. As with many arcade games, the user's best score is shown alongside the current score.

> The game is won when a tile with a value of 2048 appears on the board, hence the name of the game. After reaching the 2048 tile, players can continue to play (beyond the 2048 tile) to reach higher scores. When the player has no legal moves (there are no empty spaces and no adjacent tiles with the same value), the game ends.

> The simple gameplay mechanics (just four directions) allowed it to be used in a promo video for the Myo gesture control armband, the availability of the code underneath allowed it to be used as a teaching aid for programming, and the second-place winner of a coding contest at Matlab Central Exchange was an AI system that would play 2048 on its own.

```python
class Board():
    
    def __init__(self, board_size=4):
        
        if board_size < 4:
            board_size = 4
        
        self.board_size = board_size
        
        # Initialize the Board
        self.cells = [ [0] * board_size ] * board_size
        
    def padding(self, num):
        """
        Padding numbers with a '-' symbol so that when printing out
        the board, it is evenly spaced.

        Parameters
        ----------
        num : int
            The number to pad.
        """
        return str(num).rjust(3, '_')

    def print_board(self):
        """
        Prints out the board

        """
        for lst in self.cells:
            print([self.padding(x) for x in lst])
    
    def add_rand_tile(self):
        """
        Adds a random tile to the board if there are any available
        tiles. There is a 90% chance that the next tile is a '2'
        and a 10% chance that the next tile is a '4'.
        
        Returns
        -------
        A copy of the cell with a newly added tile, if possible.
        """
        cells = self.cells.copy()
        row_end = len(cells)
        col_end = len(cells[0])
        free_tiles = []
        
        for row_i in range(0, row_end):
            for col_j in range(0, col_end):
                if cells[row_i][col_j] == 0:
                    free_tiles.append([row_i, col_j])
                    
        if len(free_tiles) == 0:
            return cells
        
        x, y = random.choice(free_tiles)
        
        prob = random.uniform(0, 1)
        
        if prob < 0.9:
            cells[x][y] = 2
            print((x, y, 2))
            
        else:
            cells[x][y] = 4
            print((x, y, 4))
            
        return cells

    def can_move_right(self):
        """
        Check if the board can move right.
        
        Returns
        -------
        True if the board can move right, false if otherwise.
        """
        cells = self.cells
        row_end = len(cells)
        col_end = len(cells[0])

        for row_i in range(0, row_end):

            for col_j in range(0, col_end - 1):

                curr_num = cells[row_i][col_j]
                right_num = cells[row_i][col_j + 1]

                if curr_num != 0 and right_num == 0:
                    return True

                if curr_num == right_num != 0:
                    return True

        return False

    def can_move_left(self):
        """
        Check if the board can move left.
        
        Returns
        -------
        True if the board can move left, false if otherwise.
        """
        cells = self.cells
        row_end = len(cells)
        col_end = len(cells[0])

        for row_i in range(0, row_end):

            for col_j in range(col_end - 1, 0, -1):

                curr_num = cells[row_i][col_j]
                left_num = cells[row_i][col_j - 1]

                if left_num == 0 and curr_num != 0:
                    return True

                if left_num == curr_num != 0:
                    return True

        return False

    def can_move_up(self):
        """
        Check if the board can move up.
        
        Returns
        -------
        True if the board can move up, false if otherwise.
        """
        cells = self.cells
        row_end = len(cells)
        col_end = len(cells[0])

        for col_j in range(0, col_end):

            for row_i in range(row_end - 1, 0, -1):

                curr_num = cells[row_i][col_j]
                top_num = cells[row_i - 1][col_j]

                if curr_num != 0 and top_num == 0:
                    return True

                if curr_num == top_num != 0:
                    return True

        return False

    def can_move_down(self):
        """
        Check if the board can move down.
        
        Returns
        -------
        True if the board can move down, false if otherwise.
        """
        cells = self.cells
        row_end = len(cells)
        col_end = len(cells[0])

        for col_j in range(0, col_end):

            for row_i in range(0, row_end - 1):

                curr_num = cells[row_i][col_j]
                bottom_num = cells[row_i + 1][col_j]

                if curr_num != 0 and bottom_num == 0:
                    return True

                if curr_num == bottom_num != 0:
                    return True

        return False
    
    def shift_row_right(self, row):
        """
        Move all integers to the right for a given row,
        ignoring the tiles with a '0'.
        
        Returns
        -------
        Returns a copy of the row that is shifted right.
        """
        size = len(row)
        copy = list(filter(lambda x: (x != 0), row))
        copy = [0] * (size - len(copy)) + copy

        return copy

    def shift_row_left(self, row):
        """
        Move all integers to the left for a given row,
        ignoring the tiles with a '0'.
        
        Returns
        -------
        Returns a copy of the row that is shifted left.
        """
        size = len(row)
        copy = list(filter(lambda x: (x != 0), row))
        copy = copy + [0] * (size - len(copy))

        return copy

    def combine_row_left(self, row):
        """
        Adds the integers together if a pair of numbers are the same
        going in the left direction.
        
        Returns
        -------
        Returns a copy of the row with integers added together in the
        left direction.
        """
        row = row.copy()
        curr = 0
        end = len(row)

        while curr < end - 1:

            curr_num = row[curr]
            right_num = row[curr + 1]

            if curr_num == right_num != 0:
                row[curr] = curr_num + curr_num
                row[curr + 1] = 0
                curr += 1

            curr += 1

        return row

    def combine_row_right(self, row):
        """
        Adds the integers together if a pair of numbers are the same
        going in the right direction.
        
        Returns
        -------
        Returns a copy of the row with integers added together in the
        right direction.
        """
        row = row.copy()
        curr = len(row) - 1

        while curr > 0:

            curr_num = row[curr]
            left_num = row[curr - 1]

            if left_num == curr_num != 0:
                row[curr] = curr_num + curr_num
                row[curr - 1] = 0
                curr -= 1

            curr -= 1

        return row

    def move_right(self):
        """
        Moves the entire board to the right.
        
        Returns
        -------
        Returns a copy of the board that is shifted to the right.
        """
        cells = self.cells.copy()
        row_end = len(cells)
        col_end = len(cells[0])

        for row_i in range(0, row_end):

            new_row = self.shift_row_right(cells[row_i])

            new_row = self.combine_row_right(new_row)

            cells[row_i] = self.shift_row_right(new_row)

        return cells

    def move_left(self):
        """
        Moves the entire board to the left.
        
        Returns
        -------
        Returns a copy of the board that is shifted to the left.
        """
        cells = self.cells.copy()
        row_end = len(cells)
        col_end = len(cells[0])

        for row_i in range(0, row_end):

            new_row = self.shift_row_left(cells[row_i])

            new_row = self.combine_row_left(new_row)

            cells[row_i] = self.shift_row_left(new_row)

        return cells

    def move_up(self):
        """
        Moves the entire board to the up.
        
        Returns
        -------
        Returns a copy of the board that is shifted upwards.
        """
        cells = self.cells.copy()
        row_end = len(cells)
        col_end = len(cells[0])

        for col_j in range(0, col_end):

            new_col = [cells[row_i][col_j] for row_i in range(0, row_end)]

            new_col = self.shift_row_left(new_col)

            new_col = self.combine_row_left(new_col)

            new_col = self.shift_row_left(new_col)

            for row_i, num in zip(range(0, row_end), new_col):

                cells[row_i][col_j] = num

        return cells

    def move_down(self):
        """
        Moves the entire board to the down.
        
        Returns
        -------
        Returns a copy of the board that is shifted downwards.
        """
        cells = self.cells.copy()
        row_end = len(cells)
        col_end = len(cells[0])

        for col_j in range(0, col_end):

            new_col = [cells[row_i][col_j] for row_i in range(0, row_end)]

            new_col = self.shift_row_right(new_col)

            new_col = self.combine_row_right(new_col)

            new_col = self.shift_row_right(new_col)

            for row_i, num in zip(range(0, row_end), new_col):

                cells[row_i][col_j] = num

        return cells
    
    def move(self, direction):
        """
        Calls the appropriate move function given a direction.
        
        Returns
        -------
        Returns True if the board moved, False otherwise. 
        """
        if direction == 'w' and self.can_move_up():
            self.cells = self.move_up()
            return True

        if direction == 'a' and self.can_move_left():
            self.cells = self.move_left()
            return True

        if direction == 's' and self.can_move_down():
            self.cells = self.move_down()
            return True

        if direction == 'd' and self.can_move_right():
            self.cells = self.move_right()
            return True

        return False
            

class Game2048():
        
    def __init__(self):
        # TODO:
        return
        
    def get_next_event(self):
        """
        Gets a valid event from the user.
        
        Returns
        -------
        Returns the valid event for handling. 
        """
        valid_events = ['w', 'a', 's', 'd', 'q']

        while True:

            next_event = input('\nEnter Key: ')
            next_event = next_event.lower()

            if next_event in valid_events:
                break

        return next_event
            
    
    def play(self):
        """
        The main function that is called to play the 2048 game.
        """
        message = '''Use the following keys:
        
               w (UP)
    a (LEFT)   s (DOWN)   d (RIGHT)

    q - QUIT
        '''
        
        print(message)
        board = Board()
        
        # Add two random tiles to start with
        board.add_rand_tile()
        board.add_rand_tile()
        
        board.print_board()
        
        directions = ['w', 'a', 's', 'd']
        controls = ['q']
        
        game_over = False
        
        while not game_over:
            
            event = self.get_next_event()
            
            if event in directions:
                
                # Only add a Random Tile if the board moved.
                if board.move(event) == True:
                    board.add_rand_tile()
                                
                board.print_board()
                
                if (not board.can_move_left() and not board.can_move_right() and
                    not board.can_move_up() and not board.can_move_down()):
                    game_over = True
                        
            else:
                
                if event == 'q':
                    game_over = True
                    
        print('GAME OVER')
        return   
```

```python
def move_right(cells):
    
    cells = cells.copy()
    row_end = len(cells)
    col_end = len(cells[0])
    
    for row_i in range(0, row_end):
        
        new_row = shift_row_right(cells[row_i])
        
        new_row = combine_row_right(new_row)
            
        cells[row_i] = shift_row_right(new_row)
    
    return cells

def move_left(cells):
    
    cells = cells.copy()
    row_end = len(cells)
    col_end = len(cells[0])
    
    for row_i in range(0, row_end):

        new_row = shift_row_left(cells[row_i])

        new_row = combine_row_left(new_row)

        cells[row_i] = shift_row_left(new_row)
    
    return cells

def move_up(cells):
    
    cells = cells.copy()
    row_end = len(cells)
    col_end = len(cells[0])
    
    for col_j in range(0, col_end):
        
        new_col = [cells[row_i][col_j] for row_i in range(0, row_end)]
        
        new_col = shift_row_left(new_col)
        
        new_col = combine_row_left(new_col)
        
        new_col = shift_row_left(new_col)
        
        for row_i, num in zip(range(0, row_end), new_col):
            
            cells[row_i][col_j] = num
            
    return cells

def move_down(cells):
    
    cells = cells.copy()
    row_end = len(cells)
    col_end = len(cells[0])
    
    for col_j in range(0, col_end):
        
        new_col = [cells[row_i][col_j] for row_i in range(0, row_end)]
        
        new_col = shift_row_right(new_col)
        
        new_col = combine_row_right(new_col)
        
        new_col = shift_row_right(new_col)
        
        for row_i, num in zip(range(0, row_end), new_col):
            
            cells[row_i][col_j] = num
            
    return cells
```
